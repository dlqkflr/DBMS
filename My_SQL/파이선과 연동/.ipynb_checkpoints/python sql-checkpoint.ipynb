```python
#sakila db 
# 월별 렌탈 횟수가 30회 이상인 유저
# 유저 이름, 유저 이메일

db= MySQLdb.connect(
 'localhost',
'root',
'1234',
    'sakila',
charset='utf8')
```

```python
cus_df = pd.read_sql('select * from customer',db)
ren_df = pd.read_sql('select * from rental',db)
```

```python

cus_df = cus_df[['customer_id','first_name','email']]
cus_df.head()
```

```python

ren_df = ren_df[['rental_id','customer_id']]
ren_df.head()
```

```python
final_df = pd.merge(cus_df, ren_df, how='inner', on= 'customer_id')
final_df
```

```python
df30 = final_df.groupby('customer_id')['rental_id'].count()>=30
df30
```

```python
idx30 = [idx for idx, i in enumerate(df30) if i == True]
idx30
```

```python
len(idx30)
```

```python
cus_df.iloc[idx30].head()
```

```python
sql = '''
select first_name, email, count(r.rental_id) num_rent
from customer c join rental r
on c.customer_id = r.customer_id
group by c.customer_id
having num_rent >= 30
order by num_rent desc;
'''

final = pd.read_sql(sql,db)
final
```

```python
# user 별 매출
# customer, payment
# customer_id , 매출 내림차순
```

```python
c_df = pd.read_sql('select * from customer',db)
p_df = pd.read_sql('select * from payment',db)
```

```python

c_df = c_df[['first_name','customer_id']]
c_df.head(1)
```

```python
pay= p_df.groupby('customer_id')['amount'].sum()
pay_df = pd.DataFrame(pay)
pay_df
```

```python
final_df = pd.merge(c_df,pay_df, on= 'customer_id')
final_df.sort_values(['amount'], ascending = False)
```

```python
sql ='''
select c.customer_id, sum(p.amount) sell from customer c
join payment p on c.customer_id = p.customer_id
group by c.customer_id
order by sell desc;
'''

pd.read_sql(sql,db)
```
