```python
sql = '''
select left(rental_date,7) ,count(rental_id) from rental
group by left(rental_date,7);
'''

pd.read_sql(sql,db)
```

```python
#월별 전체 매출, 평균매출, 결제 갯수
```

```python
pay_df = pd.read_sql('select * from payment',db)
pay_df.head()
```

```python
pay_df['month'] = pay_df['payment_date'].apply(lambda x : str(x)[:7])
pay_df.head()
```

```python
pay_df = pay_df.drop(['staff_id','rental_id','last_update'],axis=1)
pay_df.head()
```

```python
a = pay_df.groupby('month')['payment_id'].count()
a
```

```python
b = pay_df.groupby('month').sum()['amount']
b
```

```python
c =pay_df.groupby('month')['amount'].mean()
c
```

```python
final_df = pd.DataFrame()
final_df['전체매출'] = b
final_df['결제횟수'] =a
final_df['월평균매출'] = c
```

```python
final_df
```

```python
payment_df = pay_df[['payment_id','amount','month']]
payment_df.head()
```

```python
import numpy as np

payment_df.groupby('month').agg({'amount' : [np.mean, np.sum], 'payment_id' : np.size})
```

```python
pay_df.head()
```

```python
sql = '''
select left(payment_date,7) month, count(payment_id),
sum(amount) total_price, avg(amount)
from payment group by month;
'''

pd.read_sql(sql, db)
```
