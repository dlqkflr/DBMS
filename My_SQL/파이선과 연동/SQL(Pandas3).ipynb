```python
# sakila db
# 렌탈 횟수가 30회 이상인 유저
# 유저 이름, 유저 이메일
# sql, pandas
# customer, rental
```

```python
db = MySQLdb.connect(
'localhost',
'root',
'1234',
'sakila',
charset='utf8')
```

```python
sql = '''
select first_name, email, count(rental_id) num_rent,c.customer_id

from customer c join rental r
on c.customer_id = r.customer_id

group by c.customer_id

having num_rent > 30

order by num_rent desc;
'''
pd.read_sql(sql, db)
```

```python
sql = '''
select * from customer;
'''

customer_df = pd.read_sql(sql,db)
customer_df.head()
```

```python
sql = '''
select * from rental;
'''

rental_df = pd.read_sql(sql,db)
rental_df.head()
```

```python
df = pd.merge(customer_df, rental_df, on = 'customer_id')
df.head(2)
```

```python
df30 = df.groupby('customer_id')['rental_id'].count()>30
df30
```

```python
idx30 = [ idx for idx, i in enumerate(df30) if i == True ]
idx30
```

```python
df = df[['first_name','email','customer_id']]
```

```python
df_123 = df.iloc[idx30]
len(df_123)
df_123
```

```python
df_30=df_30.groupby(['customer_id'])
```
